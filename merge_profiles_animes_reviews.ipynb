```python
import pandas as pd #utilisé pour manipuler et analyser des données tabulaires en Python.
from google.colab import files #permet de télécharger et charger des fichiers depuis l'environnement Google Colab.
import io #offre des fonctionnalités d'entrée/sortie pour gérer les flux de données.
import seaborn as sns #est utilisé pour créer des visualisations statistiques attrayantes.
import matplotlib.pyplot as plt #permet de créer des graphiques et des visualisations personnalisées dans Python.
```

```python
df_profiles = pd.read_csv('profiles.csv')
df_animes_reviews = pd.read_csv('animes_reviews.csv')
```

```python
df_profiles.shape
df_profiles.info()
```

```python
df_animes_reviews.shape
df_animes_reviews.info()
```

```python
merged_df = df_profiles.merge(df_animes_reviews, on='profile', how='inner')

merged_df.head()
```

```python
merged_df['birthday'] = pd.to_datetime(merged_df['birthday'], format='%b %d, %Y', errors='coerce')

today = pd.to_datetime('today')  # Get today's date
merged_df['age'] = (today - merged_df['birthday']).dt.days / 365.25  # Calculate age in years

merged_df['age'] = merged_df['age'].round()  # Round ages to whole numbers

merged_df['age'].fillna(-1, inplace=True)  # Fill missing ages with -1

merged_df.head()

```

```python
merged_df.drop('birthday', axis=1, inplace=True)  # Remove the 'birthday' column

merged_df.insert(2, 'age', merged_df.pop('age'))  # adjusting the 'age' column position

merged_df.head()
```

```python
merged_df.to_csv('profiles_animes_reviews.csv', index=False)
```
